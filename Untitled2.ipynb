```python
import requests
import re
from bs4 import BeautifulSoup
from nltk.tokenize import sent_tokenize, word_tokenize
import pandas as pd
```

```python
page = requests.get("https://fangj.github.io/friends/season/0101.html")
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
text = soup.get_text()
text = text.encode("ascii", "ignore").decode()
sentence = text.split("\n")
```

```python
names = []

names = re.findall("[^A-Za-z]\\n([A-Za-z ]+):", text, re.MULTILINE)

len(names)
```

```python
quotes = []
first = 0

for i in range(len(names)):
    quote = []
    for j in range(first, len(sentence)-1):
        if i < len(names)-1:
            if names[i+1] + ':' in sentence[j]:
                if names[i+1] == names[i]:
                    first = j + 1
                    quote.append(sentence[j].replace(names[i] + ':', '')[1:])
                    break
                else:
                    first = j
                    break

            if names[i] + ':' in sentence[j]:
                aux_str = sentence[j].replace(names[i] + ':', '')
                if len(aux_str) > 0:
                    if(aux_str[0] == ' '):
                        aux_str = aux_str[1:]
                    quote.append(aux_str)

            else:
                quote.append(sentence[j])
        else:
            quote.append(sentence[j].replace(names[i] + ':', ''))
            
    quotes.append(quote)

for i in range (len(names)):
    print(names[i] + ':')
    print(quotes[i])
```
